### data/gene_expression/batch_effects_normalized_mRNA_data 数据集

- 行 (identifiers)：每行代表一个基因，其对应的基因标识符如 100130426, 100133144 等。
- 列 (samples)：每列代表一个样本，其样本ID如 TCGA-OR-A5J1-01, TCGA-OR-A5J2-01 等。
- 单元格值：单元格中的数值表示该基因在该样本中的表达水平。数值越大，表示表达水平越高。

20,532 identifiers x 11060 samples


```python
import pandas as pd
gene_expression_raw_data = pd.read_csv('data/gene_expression/batch_effects_normalized_mRNA_data', sep='\t', index_col=0, nrows=100) # 读取前100行数据
gene_expression_raw_data.head()
```
